# 06. Plotting Choropleth

## 06.1 Class of Orbit - LEO

```python
#Create a data frame with just the Launch Mass (kg) AND Country of Operator/Owner
df_toplot_LEO_mass = df_LEO_2[['Country of Operator/Owner','Launch Mass (kg.)']]
```

```python
#Check the dataframe
df_toplot_LEO_mass.head()
```

```python
#Aaggregate the data based on the sum of --Launch Mass (kg.) so that each country only appears once in the dataframe. 
df_toplot_LEO_mass_agg = df_toplot_LEO_mass.groupby('Country of Operator/Owner').agg({'Launch Mass (kg.)': 'mean'}).reset_index()
```

```python
df_toplot_LEO_mass_agg.head()
```

```python
#Create a data frame with just the Expected Lifetime (yrs.) AND Country of Contractor
df_plot_LEO_lifetime = df_LEO_2[['Country of Contractor','Expected Lifetime (yrs.)']]
```

```python
#Check the dataframe
df_plot_LEO_lifetime.head()
```

```python
#Aaggregate the data based on the sum of --Launch Mass (kg.) so that each country only appears once in the dataframe. 
df_plot_LEO_lifetime_agg = df_plot_LEO_lifetime.groupby('Country of Contractor').agg({'Expected Lifetime (yrs.)': 'mean'}).reset_index()
```

```python
df_plot_LEO_lifetime_agg.head()
```

```python
#Plot df_toplot_LEO_mass

# Setup a folium map at a high-level zoom
map = folium.Map(location = [20, 0], zoom_start = 2) # Adjusted initial location for better visibility

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = country_geo, 
    data = df_toplot_LEO_mass_agg,
    columns = ['Country of Operator/Owner','Launch Mass (kg.)'],
    key_on = 'feature.properties.name', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'Spectral', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "Launch Mass (kg.)").add_to(map)
folium.LayerControl().add_to(map)

display(map)

#Viridis,#Cividis,#Spectral,#Rainbow,YlOrBr
```

```python
#To Save in HTML
map.save('LEO1MAP.html')
```

```python
#Plot df_plot_LEO_lifetime

# Setup a folium map at a high-level zoom
map = folium.Map(location = [20, 0], zoom_start = 2) # Adjusted initial location for better visibility

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = country_geo, 
    data = df_plot_LEO_lifetime_agg,
    columns = ['Country of Contractor','Expected Lifetime (yrs.)'],
    key_on = 'feature.properties.name', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'Spectral', fill_opacity=0.6, line_opacity=0.1,
    legend_name = 'Expected Lifetime (yrs.)').add_to(map)
folium.LayerControl().add_to(map)

display(map)

#Viridis,#Cividis,#Spectral,#Rainbow,YlOrBr
```

### For the classes LEO, MEO, and GEO, I explored two primary questions:
1. Who has the heaviest satellites orbiting the Earth, on average? 
- Analyzed using Launch Mass (kg) vs. Country of Operator/Owner.
2. Who manufactures satellites with the longest expected lifetime, on average?
- Analyzed using Expected Lifetime (yrs.) vs. Country of Contractor.

+ To address these questions, I created a separate dataframe for each class of orbit (LEO, MEO, GEO), containing the specific columns I wanted to plot. For each class, I plotted data on two maps, focusing on:
- Country of Operator/Owner vs. Launch Mass (kg)
- Country of Contractor vs. Expected Lifetime (yrs.)
+ This approach allowed for a visual representation of how different countries are leading in satellite mass and expected lifetime (on average) within each orbit class.
