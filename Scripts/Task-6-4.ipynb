## 06.1 Regression Analysis

```python
#Create a regression object  --The regression object is to fit the training set
regression1 = LinearRegression()

# --Launch Mass vs. Expected Lifetime
```

```python
#Fit the regression object onto the training set.
regression1.fit(X1_train, y1_train)

# --Launch Mass vs. Expected Lifetime
```

```python
#Predict the values of y using X
y1_predicted = regression1.predict(X1_test)

# --Launch Mass vs. Expected Lifetime
```

```python
#Create a plot that shows the regression line from the model on the test set
plot_test1 = plt
plot_test1.scatter(X1_test, y1_test, color='gray', s = 15)
plot_test1.plot(X1_test, y1_predicted, color='red', linewidth =3)
plot_test1.title('Satellite Launch Mass vs. Expected Lifetime (Test set)')
plot_test1.xlabel('Launch Mass (kg.)')
plot_test1.ylabel('Expected Lifetime (yrs.)')
plot_test1.show()

# --Launch Mass vs. Expected Lifetime
```

```python
#Create a regression object  --The regression object is to fit the training set
regression2 = LinearRegression()

# --Period vs. Expected Lifetime
```

```python
#Fit the regression object onto the training set.
regression2.fit(X2_train, y2_train)

# --Period vs. Expected Lifetime
```

```python
#Predict the values of y using X
y2_predicted = regression2.predict(X2_test)

# --Period vs. Expected Lifetime
```

```python
#Create a plot that shows the regression line from the model on the test set
plot_test2 = plt
plot_test2.scatter(X1_test, y1_test, color='gray', s = 15)
plot_test2.plot(X1_test, y1_predicted, color='red', linewidth =3)
plot_test2.title('Satellite Period vs. Expected Lifetime (Test set)')
plot_test2.xlabel('Period (minutes)')
plot_test2.ylabel('Expected Lifetime (yrs.)')
plot_test2.show()

# --Period vs. Expected Lifetime
```

# Observations

#### The line of best fit shows a slight positive correlation between Satellite Launch Mass and Satellite Expected Lifetime. 
#### However, the linear model, represented by the red regression line, doesn’t perfectly fit all the data points on the plot.
#### For many data points with a light launch mass (less than 300 kg), there are instances of both short and long expected lifetimes. 
#### Similarly, for a heavy launch mass (more than 700 kg), there are also instances of both short and long expected lifetimes.

----

#### This observation is also true for the relationship between 'Period (minutes) and Expected Lifetime (years).
#### The line of best fit shows a slight positive correlation, but the linear model doesn't perfectly fit all the data points.
#### For many data points with a shorter period, both short and long expected lifetimes are observed. 
#### Similarly, for longer periods, there are also instances of both short and long expected lifetimes.

-----

####  LEO satellites require less fuel to reach orbit; however, they may require more fuel to maintain their position within the orbit due to frequent adjustments needed to counteract atmospheric drag.

```python
#Create objects that contain the model summary statistics
#Find is the mean squared error and assign to ---rmse_1
rmse_1 = mean_squared_error(y1_test, y1_predicted)

# ---Lunch Mass (Kg.) Vs. Expected Lifetime (yrs.)
```

```python
#Find the R2 score and assign to ---r2_1
r2_1 = r2_score(y1_test, y1_predicted)

# ---Lunch Mass (Kg.) Vs. Expected Lifetime (yrs.)
```

```python
#Print the model summary statistics --To evaluate the performance of the model
print('Slope_1:' ,regression1.coef_)
print('Mean squared error_1: ', rmse_1)
print('R2 score_1: ', r2_1)

# ---Lunch Mass (Kg.) Vs. Expected Lifetime (yrs.)
```

```python
#Create objects that contain the model summary statistics
#Find is the mean squared error and assign to ---rmse_2
rmse_2 = mean_squared_error(y2_test, y2_predicted)

# ---Period (minutes) Vs. Expected Lifetime (yrs.)
```

```python
#Find the R2 score and assign to ---r2_2
r2_2 = r2_score(y2_test, y2_predicted)

# ---Period (minutes) Vs. Expected Lifetime (yrs.)
```

```python
#Print the model summary statistics --To evaluate the performance of the model
print('Slope_2:' ,regression2.coef_)
print('Mean squared error_2: ', rmse_2)
print('R2 score_2: ', r2_2)

# ---Period (minutes) Vs. Expected Lifetime (yrs.)
```

# Observations

### The Slope: 
- The slope shows a coefficient of 0, which means that as the launch mass changes, the expected lifetime remains the same.
- This is also true for the period it takes satellites to orbit the Earth; as the period changes, the expected lifetime remains unchanged.

### The Mean Squared Error:
- Launch Mass (kg) vs. Expected Lifetime (years) = 1.11
- Period (minutes) vs. Expected Lifetime (years) = 1.28
- The mean squared errors are small for both sets of variables.
- This indicates that the regression line passes closely to the observations, suggesting the regression model is a good fit and represents the data well.
- It can accurately predict the influence of launch mass and period on the expected lifetime.
- However, it's important to note that this does not account for the variance in the data.

### R2:
- Launch Mass (kg) vs. Expected Lifetime (years) = 0.13
- Period (minutes) vs. Expected Lifetime (years) = 0.00
- When we consider how the data points are spread out, accounting for variance, we see that the model is a poor fit for these values.

### After examining the summary statistics for both models, it is clear that the models are not performing well.
